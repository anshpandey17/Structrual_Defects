# Mehtods to control Randomness in the output

### Temperature
- Controls the Randomness and Creativity.
- 0 -------------> Deterministic, Foucsed and Factual.
- 1 ------------->  Balanced Creativity.
- More than 1 ---> More diverse and Less Coherent.


### top_p
- Nucleus sampling- Probability cutoff.


### top_k
- Sample from top k likely Tokens.
